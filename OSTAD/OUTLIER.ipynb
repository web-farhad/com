(3) Using IQR, detect height outliers and print them

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the CSV file into a pandas DataFrame
data = pd.read_csv(r"C:\Users\farha\Downloads\archive\weight-height.csv")
```

```python
data
```

**Histgram for weights**

```python
plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 2)
plt.hist(data['Weight'], bins=20, color='green', alpha=0.7)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Histogram of Weight')
```

**Histgram for heights**

```python
# Plot histograms for height and weight
plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)
plt.hist(data['Height'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Histogram of Height')
```

### IQR

```python
# Calculate the IQR for weight
Q1_weight = data['Weight'].quantile(0.25)
Q3_weight = data['Weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight
upper_limit_weight = Q3_weight + 1.5 * IQR_weight
lower_limit_weight = Q1_weight - 1.5 * IQR_weight
```

IQR

```python
print(IQR_weight)
```

### Upperlimit, lowerlimit


```python
print(lower_limit_weight,upper_limit_weight)
```

**Detect outliers based on weight, Q1, Q3, IQR**### Upperlimit, lowerlimit

```python
# Detect weight outliers
weight_outliers = data[(data['Weight'] > upper_limit_weight) | (data['Weight'] < lower_limit_weight)]
```

### Find the outlier records

```python
print("Weight Outliers:")
print(weight_outliers)
```

### IQR

```python
# Calculate the IQR for height
Q1_height = data['Height'].quantile(0.25)
Q3_height = data['Height'].quantile(0.75)
IQR_height = Q3_height - Q1_height
upper_limit_height = Q3_height + 1.5 * IQR_height
lower_limit_height = Q1_height - 1.5 * IQR_height
```

### IQR

```python
print(IQR_height)
```

### Upperlimit, lowerlimit

```python
print(lower_limit_height,upper_limit_height)
```

**Detect outliers based on height Q1, Q3**

```python
# Detect height outliers
height_outliers = data[(data['Height'] > upper_limit_height) | (data['Height'] < lower_limit_height)]
```

### Find the outlier records

```python
print("Height Outliers:")
print(height_outliers)
```
