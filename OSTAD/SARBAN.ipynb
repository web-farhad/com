```python
# Task 3: Recreate the boxplot using the 'tips' dataset
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Boxplot of Total Bill by Day')
plt.show()
```

```python
# Task 4: Recreate the scatter plot using the 'tips' dataset
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter plot of Tip vs Total Bill')
plt.show()
```

```python
# Task 5: Recreate the distplot using the 'titanic' dataset
plt.figure(figsize=(8, 6))
sns.distplot(titanic['age'].dropna(), kde=False, bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
# Task 8: Plot the count of males and females on the Titanic, separated by their class
plt.figure(figsize=(8, 6))
sns.countplot(x='class', hue='sex', data=titanic)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Count of Males and Females by Class on Titanic')
plt.show()
```

```python
# Task 9: Recreate the matrix plot using the 'titanic' dataset
plt.figure(figsize=(8, 6))
sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap of Titanic Dataset')
plt.show()
```
