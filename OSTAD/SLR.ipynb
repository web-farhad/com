```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load the CSV data into a DataFrame
data = pd.read_csv(r"C:\Users\farha\Downloads\2023-08-10T17-48-43.931Z-canada_per_capita_income.csv") 
```

```python
# Extract the features (years) and target variable (income)
years = data[['year']]
income = data['per capita income (US$)']
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Fit the model
model.fit(years, income)
```

```python
# Predict using the model
predicted_income = model.predict(years)
```

```python
# Plot the data and the linear regression line
plt.scatter(years, income, color='blue', label='Actual Data')
plt.plot(years, predicted_income, color='red', label='Linear Regression')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
# Print the coefficients
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_[0]}')
```
