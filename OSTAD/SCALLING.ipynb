### Read data from income.csv and scale the relevant features using standardization

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"C:\Users\farha\Downloads\scaled_income.csv")
```

```python
df
```

```python
# Create a StandardScaler object
scaler = StandardScaler()
```

```python
# Fit the StandardScaler object to the relevant features
scaler.fit(df[["Age", "Income"]])
```

```python
# Transform the relevant features using the StandardScaler object
scaled_features = scaler.transform(df[["Age", "Income"]])
```

```python
# Add the scaled features to the DataFrame
df[["Age", "Income"]] = scaled_features
```

```python
# Save the scaled DataFrame to a new file
df.to_csv("income_scaled.csv")
```
